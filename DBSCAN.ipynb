### DBSCAN (Density Based Spatial Clustering of Applications with Noise )

```python
import matplotlib.pyplot as plt
import random
import numpy as np
import pandas as pd
import math
```

#### Prepare dataset

```python
def fn(r,n):
    return [(math.cos(2*math.pi/n*x)*r+np.random.normal(-30,30),math.sin(2*math.pi/n*x)*r+np.random.normal(-30,30)) for x in range(1,n+1)]
```

```python
df=pd.DataFrame(fn(50,80))
df=df.append(fn(250,500))
df=df.append(fn(450,1000))
```

```python
for i in range(100):
    df=df.append([(np.random.randint(-500,500),np.random.randint(-500,500))])
    
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df[0],df[1])
plt.show()
```

#### Set values of epsilon and min-sample

```python
from sklearn.cluster import DBSCAN
model=DBSCAN(eps=30,min_samples=5)
model.fit(df)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df[0],df[1],c=model.labels_)
```
