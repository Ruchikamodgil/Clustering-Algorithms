### Agglomerative Hierarchical Clustering 
It merges the most similar or nearest pairs of clusters upto when all data point are merged into one.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\ASUS\Desktop\Iris.csv")
```

```python
df.head(5)
```

```python
X=df["PetalLengthCm"]
Y=df["SepalWidthCm"]
```

```python
Z = np.array(list(zip(X, Y)))
```

#### Dendrogram Representation

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(20, 14))  
plt.title("Dendrograms")  
plt.axhline(y=15,linestyle="--")
plt.show(sch.dendrogram(sch.linkage(Z, method='ward')))
```

Number of Clusters Required to classify the problem can be calculated by finding the longest vertical line 
from which no horizontal line passes in dendrogram.

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2)  
cluster.fit(Z)
```

```python
plt.scatter(df['PetalLengthCm'], df['SepalWidthCm'], c=cluster.labels_) 
```
